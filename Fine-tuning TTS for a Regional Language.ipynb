```python
import os

# Updated prepare_dataset function to avoid IndexError
def prepare_dataset(audio_dir, transcript_file):
    audio_transcripts = []
    with open(transcript_file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            # Ensure the line isn't empty and is correctly formatted
            if line.strip():
                parts = line.split(maxsplit=1)
                if len(parts) == 2:  # Ensure there's both an ID and a sentence
                    audio_id = parts[0]
                    text = parts[1].strip()

                    wav_file = os.path.join(audio_dir, audio_id + ".wav")  # Construct wav file path
                    if os.path.exists(wav_file):
                        audio_transcripts.append((wav_file, text))
                    else:
                        print(f"Warning: Audio file {wav_file} does not exist.")
    return audio_transcripts

# Path to your dataset (audio and transcriptions)
audio_dir = "/content/LibriSpeech/dev-clean"
transcript_path = "/content/transcripts.txt"

# Prepare dataset now
dataset = prepare_dataset(audio_dir, transcript_path)

# Print dataset to verify
print(dataset)
```

```python
from gtts import gTTS
import os

# Define your sentences from the transcript
sentences = {
    "001": "This API integration uses OAuth for secure authentication.",
    "002": "CUDA enables parallel computing on GPUs for deep learning.",
    "003": "REST APIs are essential for building scalable web services.",
    "004": "The GPU memory is allocated for parallel computing tasks.",
    "005": "ਇਹ ਇੱਕ ਸਾਧਾਰਨ ਵਾਕ ਹੈ।",
    "006": "ਮੈਨੂੰ ਪੰਜਾਬੀ ਬਹੁਤ ਪਸੰਦ ਹੈ।",
    "007": "ਆਪਣੀ ਭਾਸ਼ਾ ਵਿੱਚ ਕਮਿਊਨਿਕੇਟ ਕਰਨਾ ਮਹੱਤਵਪੂਰਨ ਹੈ।",
    "008": "ਇਹ ਟੀਟੀਐਸ ਮਾਡਲ ਪੰਜਾਬੀ ਵਿੱਚ ਬੋਲ ਰਿਹਾ ਹੈ।",
    "009": "ਮੇਰੇ ਕੋਲ ਇੱਕ ਸੁੰਦਰ ਪਹਾੜੀ ਹੈ।"
}

# Function to generate and save audio files
def generate_audio(sentences, audio_dir):
    if not os.path.exists(audio_dir):
        os.makedirs(audio_dir)

    for audio_id, text in sentences.items():
        tts = gTTS(text)
        audio_path = os.path.join(audio_dir, f"{audio_id}.wav")
        tts.save(audio_path)
        print(f"Saved audio for {audio_id}: {audio_path}")

# Path where audio files will be saved
audio_dir = "/content/LibriSpeech/dev-clean"
generate_audio(sentences, audio_dir)
```

```python
from gtts import gTTS
import os
from IPython.display import Audio

# Punjabi to Roman Transliteration dictionary with phonetic improvements and handling matras and symbols
replacements = [
    # Vowels and vowel matras
    ('ਅ', 'a'), ('ਆ', 'aa'), ('ਇ', 'i'), ('ਈ', 'ee'), ('ਉ', 'u'), ('ਊ', 'oo'),
    ('ਏ', 'eh'), ('ਐ', 'ai'), ('ਓ', 'oh'), ('ਔ', 'au'),

    # Matras
    ('ਾ', 'aa'), ('ਿ', 'i'), ('ੀ', 'i'), ('ੁ', 'u'), ('ੂ', 'oo'),
    ('ੇ', 'eh'), ('ੈ', 'ai'), ('ੋ', 'o'), ('ੌ', 'au'),

    # Consonants
    ('ਕ', 'k'), ('ਖ', 'kh'), ('ਗ', 'g'), ('ਘ', 'gh'), ('ਚ', 'ch'), ('ਛ', 'chh'),
    ('ਜ', 'j'), ('ਝ', 'jh'), ('ਟ', 't'), ('ਠ', 'th'), ('ਡ', 'd'), ('ਢ', 'dh'),
    ('ਣ', 'n'), ('ਤ', 't'), ('ਥ', 'th'), ('ਦ', 'd'), ('ਧ', 'dh'), ('ਨ', 'n'),
    ('ਪ', 'p'), ('ਫ', 'ph'), ('ਬ', 'b'), ('ਭ', 'bh'), ('ਮ', 'm'), ('ਯ', 'y'),
    ('ਰ', 'r'), ('ਲ', 'l'), ('ਵ', 'v'), ('ਸ਼', 'sh'), ('ਸ', 's'), ('ਹ', 'h'),
    ('ੜ', 'rrh'),

    # Nasalization (Bindi and Tippi)
    ('ੰ', 'n'),  # Bindi for nasalization
    ('ਂ', 'n'),  # Tippi for nasalization

    # Doubling of consonants (Addak)
    ('ੱ', 'a'),   # Addak usually doubles the consonant; handled in text

    # Nuktas
    ('਼', ''),   # Nuktas (ignored for now, can be handled if needed)
]

# Function to transliterate Punjabi text to Romanized form with phonetic improvements
def transliterate(text):
    for punjabi, roman in replacements:
        text = text.replace(punjabi, roman)

    # Handle doubling of consonants with Addak manually
    text = text.replace('ੱk', 'kk')
    text = text.replace('ੱg', 'gg')
    text = text.replace('ੱch', 'cch')
    text = text.replace('ੱp', 'pp')
    text = text.replace('ੱt', 'tt')

    return text

# Function to convert text to speech using gTTS with an Indian English accent
def text_to_speech(text, lang='en-in'):
    # Add some pauses for a more natural flow
    text = text.replace('।', '. ')
    text = text.replace(',', ', ')

    tts = gTTS(text=text, lang=lang, slow=False)
    tts.save("output.wav")  # Save as WAV file instead of MP3
    return Audio("output.wav", autoplay=True)  # Return the audio for playback in Colab

# Punjabi text input with additional pauses
# punjabi_text="ਆਪਣੀ ਭਾਸ਼ਾ ਵਿੱਚ ਕਮਿਊਨਿਕੇਟ ਕਰਨਾ ਮਹੱਤਵਪੂਰਨ ਹੈ। ਮੇਰੇ ਕੋਲ ਇੱਕ ਸੁੰਦਰ ਪਹਾੜੀ ਹੈ। ਆਪਣੀ ਭਾਸ਼ਾ ਵਿੱਚ ਕਮਿਊਨਿਕੇਟ ਕਰਨਾ ਮਹੱਤਵਪੂਰਨ ਹੈ। ਮੈਨੂੰ ਪੰਜਾਬੀ ਬਹੁਤ ਪਸੰਦ ਹੈ।"
punjabi_text = "ਮੇਰਾ ਨਾਮ ਵਿਸ਼ਾਲ ਹੈ|ਮੈਂ ਕੋਡ ਲਿਖਦਾ ਹਾਂ।"

# Transliterate to Romanized form
romanized_text = transliterate(punjabi_text)
print(f"Romanized Text: {romanized_text}")

# Convert to speech and play
text_to_speech(romanized_text)
```
